# 01 — Data Exploration

Explore the raw FIFA dataset, check distributions, and identify data quality issues.

**Goal:** Understand what we have before building anything.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='whitegrid')
%matplotlib inline
```

## 1. Load Data

```python
# Download from: https://www.kaggle.com/datasets/stefanoleone992/fifa-22-complete-player-dataset
df = pd.read_csv('../data/raw/fifa_players.csv', low_memory=False)
print(f'Shape: {df.shape}')
df.head(3)
```

## 2. Data Quality Check

```python
# Missing values
null_pct = df.isnull().mean().sort_values(ascending=False)
null_pct[null_pct > 0].head(20)
```

## 3. Overall Rating Distribution

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

rating_col = 'overall' if 'overall' in df.columns else 'overall_rating'
df[rating_col].hist(bins=30, ax=axes[0], color='steelblue', edgecolor='white')
axes[0].set_title('Overall Rating Distribution')
axes[0].set_xlabel('Rating')

# Position distribution
pos_col = 'player_positions' if 'player_positions' in df.columns else 'position'
if pos_col in df.columns:
    df[pos_col].str.split(',').str[0].str.strip().value_counts().head(15).plot(
        kind='bar', ax=axes[1], color='steelblue'
    )
    axes[1].set_title('Position Distribution')
    axes[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
```

## 4. Performance Label Distribution

Define the 4-class target and check balance.

```python
rating_col = 'overall' if 'overall' in df.columns else 'overall_rating'
p50, p80, p95 = df[rating_col].quantile([0.50, 0.80, 0.95])
print(f'Thresholds → Average: <{p50:.0f}, Good: {p50:.0f}–{p80:.0f}, Excellent: {p80:.0f}–{p95:.0f}, World Class: >{p95:.0f}')

def assign_label(r):
    if r >= p95: return 'World Class'
    if r >= p80: return 'Excellent'
    if r >= p50: return 'Good'
    return 'Average'

df['performance_label'] = df[rating_col].apply(assign_label)
df['performance_label'].value_counts().plot(kind='bar', color=['#9e9e9e','#42a5f5','#66bb6a','#ffa726'])
plt.title('Performance Label Distribution')
plt.xticks(rotation=0)
plt.tight_layout()
```

## 5. Key Correlations

```python
numeric_cols = df.select_dtypes(include='number').columns
corr = df[numeric_cols].corr()[rating_col].sort_values(ascending=False)
corr.drop(rating_col).head(15).plot(kind='bar', figsize=(10,4), color='steelblue')
plt.title(f'Correlation with {rating_col}')
plt.tight_layout()
```
